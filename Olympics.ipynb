```python
import pandas as pd
```

```python
df = pd.read_csv("summer.csv")
```

```python
#All the logics are written in readme file in github repository
```

```python
#In how many cities Summer Olympics is held so far


print("The Olympics have been held in the following ",len(df["City"].unique()),"cities-")

for city in df["City"].unique():
    print(city)
```

```python
#Which sport is having most number of Gold Medals so far

gold=[]

for i in df["Sport"].unique():
    count=0
    df2=(df[df["Sport"]==i])
    for medal in df2["Medal"]:
        if(medal=="Gold"):
            count += 1
    gold.append([i,count])

gold = pd.DataFrame(gold, columns=["Sport","Gold Medals"])
gold=gold.sort_values(by="Gold Medals", ascending = False).head()

print("The sports with most number of gold medals are:")
gold.plot(x="Sport",y="Gold Medals",kind="bar",figsize=(5,10))
gold
```

```python
#Which sport is having most number of medals so far

tot_medals=[]

for sport in df["Sport"].unique():
    count=len(df[df["Sport"]== sport])
    tot_medals.append([sport,count])

tot_medals= pd.DataFrame(tot_medals, columns=["Sport","Medals"])
tot_medals= tot_medals.sort_values(by="Medals", ascending = False).head()

print("The sports with most number of medals are:")
tot_medals.plot(x="Sport",y="Medals",kind="bar",figsize=(5,15))
tot_medals
    
```

```python
#Which player has won most number of medals

ply_medals=[]

for player in df["Athlete"].unique():
    count=len(df[df["Athlete"]== player])
    ply_medals.append([player,count])

ply_medals= pd.DataFrame(ply_medals, columns=["Athlete","Medals"])
ply_medals= ply_medals.sort_values(by="Medals", ascending = False).head()

print("The Athletes with most number of medals are:")
ply_medals.plot(x="Athlete",y="Medals",kind="bar",figsize=(5,5))
ply_medals
```

```python
#Which player has won most number Gold Medals of medals

ply_gold=[]

for i in df["Athlete"].unique():
    count=0
    df2=(df[df["Athlete"]==i])
    for medal in df2["Medal"]:
        if(medal=="Gold"):
            count += 1
    ply_gold.append([i,count])

ply_gold = pd.DataFrame(ply_gold, columns=["Athlete","Gold Medals"])
ply_gold=ply_gold.sort_values(by="Gold Medals", ascending = False).head()

print("The athlete with most number of gold medals are:")
ply_gold.plot(x="Athlete",y="Gold Medals",kind="bar")
ply_gold
```

```python
#In which year India won first Gold Medal in Summer Olympics


for i in df["Country"].unique():
    if i == "IND":
        df2=df[df["Country"]=="IND"]
        for medal in df2["Medal"]:
            if medal=="Gold":
                df3 = df2[df2["Medal"]=="Gold"]
                
df3=df3.head(1)
print("India won its first gold in")
df3
        
```

```python
#Which event is most popular in terms on number of players

events=[]

for i in df["Event"].unique():
    df2=df[df["Event"]==i]
    events.append([i,len(df2["Athlete"].unique())])
    
events = pd.DataFrame(events, columns=["Event","Athletes"])
events=events.sort_values(by="Athletes", ascending = False).head()

events.plot(x="Event",y="Athletes",kind="bar")
events
    
```

```python
#Which sport is having most female Gold Medalists

female=[]

for i in df["Gender"].unique():
    if i == "Women":
        df2=df[df["Gender"]=="Women"]
        for medal in df2["Medal"]:
            if medal=="Gold":
                df3 = df2[df2["Medal"]=="Gold"]
                
for j in df3["Sport"].unique():
    count=len(df3[df3["Sport"]== j])
    female.append([j,count])
    
female = pd.DataFrame(female, columns=["Sport","Female gold Medalist"])
female = female.sort_values(by="Female gold Medalist", ascending = False).head()

female.plot(x="Sport",y="Female gold Medalist",kind="bar")
female
    
```
